```python
# Analyze the “auto mpg data” and draw a pair plot using seaborn library for mpg, weight, and origin.
# This dataset was taken from the StatLib library maintained at Carnegie Mellon University. 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_auto_dataset = pd.read_csv('auto-mpg.csv')
```

```python
df_auto_dataset.head()
```

```python
# Write a user defined function for origin 1-USA, 2-Europe, 3-Asia
def origin(num):
    if num==1:
        return 'USA'
    elif num==2:
        return 'Europe'
    else:
        return 'Asia'

df_auto_dataset['origin'] = df_auto_dataset['origin'].apply(origin)
```

```python
df_auto_dataset.head(30)
```

```python
# Draw the pair plot using sns for mpg, weight, origin and with hue origin, set the size to 4.
sns.pairplot(df_auto_dataset[['mpg', 'weight', 'origin']], hue='origin', size=4)
```

```python
# Data from Ohio state, leading cause of death in the state from year 2012

cause = 'Chronic Diseases', 'Unintentional Injuries', 'Alzheimers', 'Influenza and Pneumonia', 'Sepsis', 'Others'
percentile = [62,5,4,2,1,26]
```

```python
plt.figure(figsize=(10,10))

# explode the largest pie in the data set
explode = (0.05,0,0,0,0,0)
plt.pie(percentile, labels=cause, explode=explode, autopct='%1.1f%%', startangle=70)

# set axis equal to draw pie as circle 
plt.axis('equal')
plt.title('Ohio State - 2012: Leading Causes of Death')
plt.show()
```
